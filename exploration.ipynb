#### convert duration in a seconds to  duration in a minutes

```python
df2['duration_sec'] = df2['duration_sec']/60
```

```python
df2.rename(columns={'duration_sec' : 'duration_min'},inplace=True)
```

#### convert start_time to hours

```python
df2['hour_of_day'] = pd.DatetimeIndex(df2['start_time']).hour
```

# Univariate Exploration


```python
bins = np.arange(0, df2['duration_min'].max()+1, 1)
plt.figure(figsize=[10, 6])
plt.title('Distribution of Duration in minutes',fontsize=15)
plt.hist(data=df2, x='duration_min', bins=bins)
plt.xlabel('Duration in a minutes',fontsize=15)
plt.ylabel('Count',fontsize=15);
plt.xlim(0, 200)
```

### How long do people spend the trip in minutes ?
#### from the graph we can see some outliers at the end to extremely skewed the data. It's right-skewed we can see that the peak usage falls in that 2 - 20 minutes range, so people are using these bikes for quick rides.

# Bivariate Exploration


```python
plt.figure(figsize=(10,6))
sb.countplot(data = df2, x ='user_type', hue = 'member_gender', palette='icefire_r');
plt.title('Number of rides of each gender',fontsize=15);
plt.xlabel('User type',fontsize=15)
plt.ylabel('Count',fontsize=15);
```

#### From the graph, we can see males users are much more than females and others in both types customers and subscribers

```python
plt.figure(figsize=(15,8))
sb.countplot(data=df2, x='hour_of_day', hue='user_type', palette='rocket' );
plt.xlabel('Hours in the day',fontsize=15)
plt.ylabel('Count',fontsize=15);
plt.title('Hours in the day Distribution', fontsize=15)
plt.legend(title=None);
```

#### From the graph, we can see the total number of the trip is very much bigger for members for each kind of period of time

```python
plt.figure(figsize=(10,6))
sb.barplot(x=df2['member_gender'],y=df2['duration_min'], color = sb.color_palette()[3])
plt.ylabel('Duration average',fontsize=15)
plt.xlabel('Member gender',fontsize=15)
plt.title('Duration average for different genders',fontsize=15);
```

#### From the graph, we can see that although males and females are much more than others, but their use of the bike is less than others.

```python
plt.figure(figsize=(10,6))
sb.barplot(x=df2['user_type'],y=df2['duration_min'], color = sb.color_palette()[4])
plt.ylabel('Duration average',fontsize=15)
plt.xlabel('User types',fontsize=15)
plt.title('Duration average for different genders',fontsize=15);
```

#### From the graph, we can see that although subscribers more than customers, but the customers use the bike double than subscribers.
